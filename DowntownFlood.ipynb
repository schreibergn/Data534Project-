```python
#splitting data into train and test df because they were not split originally
df_floodtrain = transformed_dfwork2.sample(frac = 0.8,random_state = 200)
df_floodtest = transformed_dfwork2.sample(frac = 0.2, random_state = 200)
df_floodtrain.info()
df_floodtest.info()
```

```python
#Random Forest model
X = df_floodtrain.drop('REASON_FLOOD', axis = 1)
y = df_floodtrain['REASON_FLOOD']

from sklearn.ensemble import RandomForestClassifier

cls = RandomForestClassifier(random_state=44)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)

cls.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

#Training error
y_pred = cls.predict(X_train)
print(accuracy_score(y_train, y_pred))

#Validation(testing) error
y_pred = cls.predict(X_test)
print(accuracy_score(y_test, y_pred))

#y_pred = cls.predict(df_floodtest)
#pd.DataFrame({})
#need to figure out how to test with test df
```

```python
from sklearn.metrics import confusion_matrix
 
cm = confusion_matrix(y_test,y_pred)
print(cm)
#this is the total number of correct predictions beacuse the accuracy is 100%
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
 
X = df_floodtrain.drop('REASON_FLOOD', axis = 1)
y = df_floodtrain['REASON_FLOOD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
 
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
 
color = 'white'
matrix = plot_confusion_matrix(knn, X_test, y_test, cmap=plt.cm.Blues)
matrix.ax_.set_title('Confusion Matrix', color=color)
plt.xlabel('Predicted Label', color=color)
plt.ylabel('True Label', color=color)
plt.gcf().axes[0].tick_params(colors=color)
plt.gcf().axes[1].tick_params(colors=color)
plt.show()
```

```python
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

cmap = LinearSegmentedColormap.from_list('custom blue', [(0, '#c9c9f5'), (0.2, '#4946fa'), (1, '#051ba8')], N=80)

x = MergedHomeFlood['RoadClosureDate']
y = MergedHomeFlood['sold_price']
colors = (y - y.min()) / (y.max() - y.min())

red_cmap = plt.cm.get_cmap('Reds')
red_norm = plt.Normalize(vmin=y.min(), vmax=y.max())
red_colors = red_cmap(red_norm(y))

colors = np.zeros((len(x), 4))
for i in range(len(x)):
    if x[i] < 10:
        colors[i] = red_colors[i]
    else:
        colors[i] = cmap(x[i])


plt.figure(figsize=(15, 15))
plt.scatter(x, y, c=colors, s=100)
plt.colorbar(label='Median Sale Price')

plt.title('Housing Value Comparison with Flood Rate', fontsize=18, y=1.03)
plt.xlabel('Number of Street Closures', fontsize=14, labelpad=15)
plt.ylabel('Median Sale Price', fontsize=14, labelpad=15)
plt.tick_params(labelsize=12, pad=6)

plt.show()
```
